### Lists

Lists are collections of values (of any format). These are useful to store information. 
Adding or multiplying lists together _doesn't_ add the values, it makes a bigger list. 

Note that counting in python starts from zero

```python
a = [1,2,3,4,5]
b = [6,7,8,9,10]

c = a+b
print('a+b where a and b are lists = ',c)
print('fifth element in c is',c[4])
print('the number of data points in c is',len(c))
```

```python
np.array([1.,2.,8]) + np.array([1.])
```

### Arrays

Arrays are another way of collecting data. For these, the values must all be of the same type. Arrays can be multi-dimensional. Adding arrays together will add the values. 

Useful functions include `np.arange()` to make an array of values and `np.zeros()` which makes an array of zeros

```python
np.array([1.,2.,8]) + np.array([10.])
```

```python
# Generate an array of values from 4 to 7 in steps of 0.5. Note that the right-hand value is omitted
np.arange(4,7,0.5)
```

### Slicing and obtaining subsets of arrays

We can obtain a subset of an array using: 

 * `a[start:stop]`  ----  items from start through stop-1
 * `a[start:] ` ---- items start through the rest of the array
 * `a[start:stop:step]` ---- can have a step size to pull out say every 100 points from the array

Note `start`, `stop` and `step` must all be integers

**NB** numpy also has a range of functions for calculating key statistics (e.g. `np.mean()`,`np.max()`,`np.std()`)

```python
y = np.arange(0,10,1)
print(y[3:7])
print(np.mean(y[3:7]))
```

### Iterating over values - for loops

Python can iterate over the items in a list. For loops are good if you want to repeat something a fixed number of times. We can use the `range` function to give a set of integers to loop over. e.g. `range(6,10,2)` will give a range of 6-10 in steps of 2.

The for loop needs a colon (:) and everything after that (related to the for loop) must be indented.

We can save the answers by using the `append()` function

```python
y = []
for k in range(1, 10):
    y.append(2*k)
print(y)
```

#### Factorising code
A much simpler alternative way of performing a calculation multiple times is shown below:

```python
y2 = [2*k for k in range(0,10)]
print(y2)
```

**Warning**: when performing calculations, always make sure that your values are of the expected type

### Conditional (if; else) Statements

if statements allow us to deal will different conditions. 

```python
y = np.arange(0,5)

for val in y: 
    if val%2==0:
        print(f'{val} is an even number')
    else:
        print(f'{val} is odd')
```

### Equations and functions

If a piece of code requires the same calculation multiple times it's best to define it as a function.

```python
def my_formula(x):
    return x**2+2

a  = 3.4
print(my_formula(a))
```

Functions can have multiple inputs. Inputs can have default values, meaning that they don't need to be provided at input

```python
def velocity(distance, time=1):
    value = distance*time
    return value
print(velocity(10))
```

## Plotting data

An important aspect of all data analysis is _data visualisation_. When determining a result, time should be given to the best way to convey it. 

The most commonly used library for plotting sciencitific data in Python is `matplotlib`.  There are a wealth of [online examples](https://matplotlib.org/stable/) options for this library. In this course, focus on making the scientific data appear clear through e.g. different markers, labelled axes, suitable fontsizes and legends.

`matplotlib` allows for simple plotting through the `plt.plot()`, `plt.hist()` and `plt.errorbar()` functions. For publication quality results the figure size can be defined by defining figures and axes.  

Key options: 
   * `color`: the color of the desired line or data points. Many predefined options; HEX code format is also allowed 
   * `alpha`: the transparency of the line / data points
   * `ls` : the linestyle to be plotted ('-','--','-.')
   * `lw` : the width of the line to be plotted
   * `label`: the label to use in a given legend
   * `marker` (sometimes `fmt`) : the marker type to use for data points

Note that LaTeX format can be used for labels. For a full list of options see [here](https://matplotlib.org/3.3.0/api/pyplot_summary.html).

```python
# Plot the function y=sin(x) from 0 to 6*!pi
x = np.arange(0,6*np.pi,0.1*np.pi)
plt.plot(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# We can also plot data points (with or without an error)
x_data = np.random.uniform(0,18,100)
y_data = np.sin(x_data) + np.random.normal(-0.1,0.1,100)
y_err  = np.random.uniform(0.05,0.2,100)
plt.errorbar(x_data,y_data, fmt='o', color='blue')
plt.errorbar(x_data,y_data, y_err, fmt='.', color='orange')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# For a more extensive example let's plot the difference between a y=sin() and y=cos() function. 
# We can define the figure environment to plot into, and split it in to multiple panels
x_vals = np.arange(0,6*np.pi,0.1*np.pi)
fig   = plt.figure(figsize=[10,4])
grid  = fig.add_gridspec(ncols=1, nrows=2, width_ratios=[1], height_ratios=[1, 0.5], wspace=0.0, hspace=0.1)
ax1   = fig.add_subplot(grid[0,0])
_     = ax1.plot(x_vals, np.sin(x_vals), color='magenta', label='y=sin(x)')
_     = ax1.plot(x_vals, np.cos(x_vals), color='dodgerblue', label='y=cos(x)')
_     = ax1.set_xlim(0,5*np.pi)
_     = ax1.legend(fancybox=True, framealpha=1.0, loc='upper right', fontsize=12)
_     = ax1.set_ylabel('y', fontsize=14)
_     = ax1.get_xaxis().set_visible(False)
ax2   = fig.add_subplot(grid[1,0])
_     = ax2.plot(x_vals, np.cos(x_vals)-np.sin(x_vals), color='orange', label='y=cos(x)-sin(x)')
_     = ax2.set_xlabel('x', fontsize=14)
_     = ax2.set_ylabel('$\Delta$ y', fontsize=14)
_     = ax2.legend(fancybox=True, framealpha=0.8, loc='lower right')
```
